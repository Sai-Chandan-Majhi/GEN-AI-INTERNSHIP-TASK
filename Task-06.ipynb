### Problem Statement 1:Smart Parking Lot Management System

```python
def smart_parking(parking_capacity, vehicle_logs):

    count_in = 0
    count_out = 0

    for log in vehicle_logs:
        if log == 'IN':
            count_in += 1
        elif log == 'OUT':
            count_out += 1

    parked = count_in - count_out

    if parked >= parking_capacity:
        status = 'Full'
    else:
        status = 'Available'

    print(f'Currently Parked Vehicles: {parked}')
    print(f'Parking Status: {status}')


parking_capacity = int(input('Enter Parking Capacity: '))
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(parking_capacity, vehicle_logs)
```

###  Problem Statement 2:Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    eta = distance * 5

    # traffic delay
    if traffic.lower() == "low":
        eta += 5
    elif traffic.lower() == "medium":
        eta += 8
    elif traffic.lower() == "high":
        eta += 10

    # weather delay
    if weather.lower() == "cloudy":
        eta += 3
    elif weather.lower() == "rainy":
        eta += 5

    return eta


print("Estimated Delivery Time:", estimate_delivery_time(8, "High", "Rainy"), "minutes")
```

### Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def movie_theatre(total_seats, booked_seats):

    booked_count = len(booked_seats)

    if booked_count > total_seats:
        print("Invalid booking data")
        return

    occupancy = (booked_count / total_seats) * 100
    print(f"Occupancy: {int(occupancy)}%")

    # Show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    elif occupancy >= 50:
        status = "Moderate"
    elif occupancy > 0:
        status = "Few Bookings"
    else:
        status = "Empty"

    print("Show Status:", status)

    # Suggest additional show
    if occupancy >= 85:
        print("Suggestion: Open additional show")
    else:
        print("Suggestion: No extra show needed")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

movie_theatre(total_seats, booked_seats)
```

###  Problem Statement 4:Cloud Server Load Classification System

```python
def server_load(cpu):
    avg=sum(cpu)/len(cpu)
    if avg<50:
        s='Normal'
    elif avg>=50 and avg<80:
        s='Warning'
    else:
        s='Critical'
    
    print(f'Average CPU Load: {int(avg)}%')
    print(f'Server Status: {s}')
CPU =[45, 60, 70, 85, 90]
server_load(CPU)
```

###  Problem Statement 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```
