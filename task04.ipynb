```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("❌ Minimum recharge amount is ₹50")
        elif amount not in valid_plans:
            print("❌ Invalid plan! Available plans:", valid_plans)
        else:
            print("✅ Recharge Successful!")
            break   # exit loop if valid
validate_recharge()
```

```python
def check_inventory(products):
    for product in products:
        stock = products[product]

        if stock < 15:
            print(product, ": Reorder Alert (Stock =", stock, ")")
        else:
            print(product, ": Stock OK (Stock =", stock, ")")

inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 5,
    "Monitor": 18
}

check_inventory(inventory)
```

```python
def calculate_result(marks):
    total = 0

    for mark in marks:
        total = total + mark

    average = total / len(marks)

    if average >= 50:
        return "Average = " + str(average) + " : Pass"
    else:
        return "Average = " + str(average) + " : Fail"

marks_list = [60, 55, 70, 40, 50]
result = calculate_result(marks_list)
print(result)
```

```python
def calculate_fare():
    base_fare = 50
    rate = 12

    while True:
        km = float(input("Enter distance in KM: "))
        peak = input("Is it peak hour? (yes/no): ")

        fare = base_fare + (km * rate)

        if peak.lower() == "yes":
            fare = fare + (fare * 0.25)

        print("Total Fare = ₹", fare)

        again = input("Calculate again? (yes/no): ")
        if again.lower() != "yes":
            break

calculate_fare()
```

```python
def check_attendance(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Attendance = " + str(percentage) + "% : Eligible"
    else:
        return "Attendance = " + str(percentage) + "% : Not Eligible"

attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]
print(check_attendance(attendance_list))
```

```python
def check_password(password):
    special = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and @#$)"

pwd = input("Enter password: ")
print(check_password(pwd))
```
