### Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def show_status(self):
        print(f"{self.light_name} is {self.status}")


room = SmartLight("Bedroom Light")
room.turn_on()
room.show_status()
```

### Problem 2: Employee ID Card System

```python
class Employee_id:
    def __init__(self,name,id,dept):
        self.name=name
        self.id=id
        self.dept=dept
    def show_result(self):
        print('Employee ID Card')
        print(f'Name: {self.name}')
        print(f'ID: {self.id}')
        print(f'Department: {self.dept}')
e1=Employee_id('Rahul','EMP102','IT')
e1.show_result()
```

### Problem 3: Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name.title()

        if len(phone) == 10 and phone.isdigit():
            self.phone = phone
        else:
            self.phone = "Invalid Number"

    def display(self):
        print("Contact Saved")
        print(f"Name : {self.name}")
        print(f"Phone: {self.phone}")


c1 = Contact("Anita", "9876543210")
c1.display()
```

### Problem 4: Product Price Tag Generator

```python
class product:
    def __init__(self,pname,price):
        self.pname=pname
        if isinstance(price, (int, float)) and price > 0:
            self.price = price
        else:
            self.price='Invalid Price'
    def show_pricetag(self):
        print(f'Product: {self.pname}')
        print(f'Price: ₹{self.price}')
p=product("Headphones",2499)
p.show_pricetag()
```

### Problem 5: Movie Rating Display System

```python
class movierating:
    def __init__(self,mname,rating):
        self.mname=mname
        if isinstance(rating,(int,float))and 0 <= rating <= 5:
            self.rating=rating
        else:
            self.rating='invalid'
    def show_details(self):
        print(f'Movie Name: {self.mname}')
        print(f'Rating: {self.rating}')
m=movierating('Inception',4.8)
m.show_details()
```
