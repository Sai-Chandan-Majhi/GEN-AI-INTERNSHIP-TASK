```python
sentence='python is easy and python is powerful'
x=sentence.split()

a=[]
for i in x:
    if i not in a:
        a.append(i)
print(f'Unique words count:{len(a)}')
print(f'Unique Words:{set(a)}')
```

```python

attendance = ["P", "P", "A", "P", "P"]

# Step 1: Count how many days the student was Present ("P")
present_count = attendance.count("P")

# Step 2: Find the total number of days
total_days = len(attendance)

# Step 3: Calculate the percentage
# Formula: (Part / Total) * 100
attendance_percentage = (present_count / total_days) * 100

print(f"Attendance Percentage: {attendance_percentage}")
```
