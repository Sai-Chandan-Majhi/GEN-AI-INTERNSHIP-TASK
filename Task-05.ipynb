```python
def analyze_engagement(likes):
    total=0
    for i in likes:
        total=total+i
        
    if total>=1000:
        status='Viral Post'
    else:
        status='Noraml Engagement'
    
    print(f'Total Likes:{total}')
    print(f'Post sattus:{status}')
    
    
post_likes=[200,300,300,250]
analyze_engagement(post_likes)
```

```python
def agriculture(rainfall):
    total=0
    for i in rainfall:
        total=total+i
    
    avg=total/len(rainfall)
    if avg>70:
        s='Adequate Rainfall'
    else:
        s='Inadequate Rainfall'
    
    
    print(f'Average Rain:{avg}')
    
    print(f'Rainfall Status: {s}')
    
x=[72,72,72,72]
agriculture(x)
```

```python
def socialmedia(account):
    if len(account)!=len(set(account)):
        s='yes'
    else:
        s='no'
        
    print(f'Duplicate Accounts Found:{s}')

    
x=['sai','elon','trump','elon']
socialmedia(x)
```

```python
def premium_threshold(price):
    premium=[]
    for i in price:
        if i>2000:
            premium.append(i)
    return premium         
    
s=[2500,400,3200]    
result=premium_threshold(s)
print(f'Premium Crops: {result}')
```

```python
def health_checker(errors):
    if errors==0:
        status='Healthy'
    elif errors<=5:
        status='Minor Issues'
    else:
        status='Critical Issues'
        
    print(f'Error Count: {errors}')
    print(f'System Status: {status}')

health_checker(7)
```

```python
def exam(eligibility):
    total_attendance=0
    
    for i in eligibility:
        total_attendance+=i
    
    
    avg_attendance=total_attendance/len(eligibility)
    
    
    if avg_attendance>75:
        status='Eligible'
    else:
        status='Not Eligible'
        
    print(f'Attendance Percentage: {avg_attendance}')
    print(f'Exam Eligibility: {status}')
    
attendance=[90,70,80,85,75]
exam(attendance)
```

```python
def bill(x):
    total = 0

    if x > 200:
        total += (x - 200) * 7
        x = 200

    if x > 100:
        total += (x - 100) * 5
        x = 100

    total += x * 3

    # usage category
    if total < 500:
        s = 'Low use'
    elif total < 1000:
        s = 'Medium use'
    else:
        s = 'High use'

    print(f'Bill Amount: {total}')
    print(f'Usage Category: {s}')


bill(500)
```
